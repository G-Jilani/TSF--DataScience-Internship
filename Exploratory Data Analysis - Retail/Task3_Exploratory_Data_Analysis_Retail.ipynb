# Observationn


*   Country has only one value, United States, we will be dropping the country column.



```python
data = df.drop(columns=['Country','Postal Code'] , axis=1)
data.head()
```

*We drop country column and postal. code column because it does not directly affect profit and sales.



# **Data Visualization**

```python
#history distribution of numeric features
data.hist(bins=20,figsize=(18,16),)
```

# **HeatMap**

```python
#correlation between features
corr = data.corr()
corr
```

```python
# Visualizing correlation between features
plt.figure(figsize=(15,9))
sns.heatmap(corr,annot=True,center=1)
plt.show()
```

# **Observation**

*   From the heatmap we can see here a positive cor-relation between sales and profit that is 0.48 , thus Profit increases ,simultanously sales increases. 
*  And here we can see a negative cor-relation between profit an discount is -0.22 and sales and discount of -0.028, thus discount increases , simultanously profit goes down .



```python
#calculating statewise profit
profit_state_g = data.groupby('State')['Profit'].sum()
profit_state_g
```

# **State-Wise Profit of US**

```python
#State-wise profit of US
plt.subplots(figsize=(15,17),)
profit_state_g.plot(kind='barh', fontsize=15,color='darkslategrey')
plt.xlabel('profit',fontsize=15,)
plt.ylabel('state',fontsize=13)
plt.title('statewise profit',fontsize=14)
plt.show()
```

# **Observation**
*   Highest profit is generated by California and Newyork.
*   Lowest profit is generated by Texas and Ohio.

**By the subplot we can clearly see that:**
*     Top 5 states giving profit are California , Newyork , Washington and Michigan .
*    Top 5 states giving loss are Texas , Ohio , lllinois , Pennsylvania



```python
#calculating statewise sales
Sales_state_g = data.groupby('State')['Sales'].sum()
Sales_state_g
```

```python
#State wise Sales of US
plt.figure(figsize=(18,12))
sns.barplot(x=Sales_state_g.index, y=Sales_state_g.values , palette='rocket')
plt.ylabel("Sales")
plt.xticks(rotation=80)
plt.show()
```

# **Observation**
*   Thus giving max profit California also has most sales.
*   We can see that the Lowest sales are from North Dakota.

*   By the subplot we can signify that:
*   Top 5 states giving most of the sales are from California , Newyork, Washington and Texas.

*   Top 5 states giving loss are North Dakota , Maine , South Dakota, West Virginica.








```python
#Comparing Sales and Profit and Statewise
sales_Profit = data.groupby('State')['Sales','Profit'].sum()
sales_Profit.plot.bar(figsize=(14,9),)
plt.show()
```

# **Observation**

*   State California, Washington, New York, Michigan have highest sales and profit.
*   State Texas, Pennsylvania, Ohio, lllinois have higher sales but profit is negative (in loss).



```python
data.Discount.head(9)
```

```python
#State wise average Discount of US
avg_dis = np.round(data.groupby('State')['Discount'].mean()*100,decimals=0)
avg_dis
```

```python
#State wise avg_discount of US
plt.figure(figsize=(15,9))
sns.barplot(x=avg_dis.index,y=avg_dis.values,palette='Set1')
plt.xticks(rotation=80)
plt.show()
```

# **Observation**

*   lllinois and Texas has highest average discount among other states of US.
*   We can also see that here Top 5 states having most discount are lllinois, Texas, Pennsylvania, Ohlo and Colorado.



```python
#Lineplot between profit and Discount
plt.figure(figsize=(8,6))
sns.lineplot(x='Discount',y='Profit',data = data,color='green')
plt.show()
```

```python
# Calculating category wise sales, Discount, Profit
cat_data = data.groupby(['Category'])[['Sales','Discount','Profit']].mean()
cat_data
```

```python
#plotting donut chart to see category wise sales, Discount, Profit
ex=[0.05,0.05,0.05]
cat_data.plot.pie(subplots=True,figsize=(19,20),autopct='%0.1f%%',label=cat_data.index,pctdistance=0.65,explode=ex)
circle = plt.Circle(xy=(0,0),radius=0.50,facecolor='white')
plt.text(0,0,'category',horizontalalignment='center',verticalalignment='center')
plt.gca().add_artist(circle)
plt.show()
```

# **Observation**

*   Technology category has the Maximum Sales and profit and less discount compare to furniture and office.
*   Office supplies and Furniture category has Minimum sales and Profit as compare to Technology category.



```python
# Calculating Ship Mode wise sales, Discount, Profit
ship_data = data.groupby(['Ship Mode'])[['Sales','Discount','Profit']].mean()
ship_data
```

```python
#plotting donut chart to see category wise sales, Discount, Profit
ex=[0.05,0.05,0.05,0.05]
ship_data.plot.pie(subplots=True,figsize=(19,20),autopct='%0.1f%%',label=cat_data.index,pctdistance=0.65,explode=ex)
#circle
circle = plt.Circle(xy=(0,0),radius=0.50,facecolor='white')
plt.text(0,0,'Ship Mode',horizontalalignment='center',verticalalignment='center')
plt.gca().add_artist(circle)
plt.show()
```

# **Observation**


*   In Ship Mode, Same Day has the Highest sales.
*   First class has the Highest Profit and Discount.
*   In Ship Mode, Satandard Class has lowest sales as well as lowest profit.






```python
#Comparing Sales Sub-Category Wise
sub_cat = data.groupby('Sub-Category')['Sales'].mean()
#Sub_cat
plt.figure(figsize=(15,9))
sns.barplot(x=sub_cat.index,y=sub_cat.values,palette='viridis')
plt.xticks(rotation=80)
plt.show()
```

# **Observation**
*   Sub-Category items Copier, Machines and Table have the highest Sales.




```python
#Comparing Sales Sub-Category Wise
sub_cat = data.groupby('Sub-Category')['Profit'].sum()
#Sub_cat
plt.figure(figsize=(15,9))
sns.barplot(x=sub_cat.index,y=sub_cat.values,palette='icefire_r')
plt.xticks(rotation=80)
plt.show()
```

# **Observation**
*   We can see here Sub-Category , items Copiers has the highest Sales as well as highest profit margin.
*   We can see here Machine had second highest sales but less profit margin, due to giving large amount of discount.
*   n sub-category, item Table had third highest sales but it's profit is in negative (on loss).
