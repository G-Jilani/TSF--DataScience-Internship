# **Missing data Analysis**

```python
missing_values = df.isnull().sum()
missing_values
```

```python
plt.rcParams["figure.figsize"]=[15,5]
sns.heatmap(df.isnull(), cbar = False)
plt.show()
```

```python
missing_values = df.isnull().sum()
total = df.isnull().sum().sort_values(ascending=False)
percent = ((df.isnull().sum()/df.shape[0])*100)
percent = percent.sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data['Type'] = df[missing_data.index].dtypes
missing_data
```

```python
nan_data=missing_data[missing_data['Percent']>60]
nan_data
```

```python
nan_data.index
```

```python
df_num=df.select_dtypes(include=np.number)
df_cat=df.select_dtypes(include=object)
```

```python
for i in df_num.columns:
    df[i].fillna(df[i].median(), inplace=True)
for i in df_cat.columns:
    df[i].fillna(df[i].mode()[0], inplace=True)
```

# **Identifying hot zones of terrorism**
A table depicting the frequencies of the two features ie., region_txt, attacktype1_txt

```python
pd.crosstab(df['region_txt'],df['attacktype1_txt'])
```

# **Inference** : From the above table we can see that Middle East and North Africa are highly affected due to Bombing and Explosion

**Forming a table that gives the number of attcks each year with respect to the region and plotting a graph for the same**

```python
pd.crosstab(df['iyear'], df['region_txt']).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

# **Inference** : The above graph depicts that South Asia, Middle East and North Africa are affected the most each year due to terrorism

**A countplot on top 10 countries affected due to terrorism**

```python
ax = sns.countplot(df['country_txt'],order = df['country_txt'].value_counts().iloc[:10].index, palette='Greens');
ax.set_xlabel('Country', fontsize=12)
ax.set_ylabel('Count', fontsize=12)
ax.set_title('Terrorism in different country- top 10',fontsize=12);
```

# **Inference** : From the above graph we can see that Iraq is the most affected due to terrorism followed by Pakistan, Afghanistan and India

# **Identifying security issues and deriving insights**
Frequency of terrorist gangs

```python
gangs=df['gname'].value_counts().reset_index()
gangs.rename(columns={"index":'Gang','gname':'Count'},inplace=True)
gangs
```

# **A bar plot on the top 10 gangs who are the reason for terrorism vs their count**

```python
sns.barplot(x='Count', y='Gang', data=gangs[:10]);
```

# Inference:  From the above bar plot, we can see that the unknown set of terrorists are the reason for terrorism maximum number of times followed by Taliban and ISIL.

# **A table showing the frequency of weapons used during the attack**

```python
weapons=df['weaptype1_txt'].value_counts().reset_index()
weapons.rename(columns={"index":'Weapons','weaptype1_txt':'Count'},inplace=True)
weapons
```

# **A bar plot on the top 5 weapons used vs their count**

```python
sns.barplot(x='Count', y='Weapons', data=weapons.iloc[:5])
```

# Inference: From the above bar plot we can see that explosives are the mostly used weapons to attack different territories followed by Firearms.

# **A bar plot depicting the terrorists most favourite to least favourite target**

```python
sns.barplot(x=df['targtype1_txt'].value_counts().values, y=df['targtype1_txt'].value_counts().index, palette='Reds');
plt.title("Terorrist's favorite target", fontsize=25);
```

# Inference: From the above bar plot, we can see that the highly targetted section for the terrorists is Private Citezens and Property followed by Military and Police

# **A table consisting of the recent 5 years with the sum of people killed and wonded in the corresponding years**

```python
affects=df[['iyear','nkill','nwound']]
affects=affects.groupby('iyear')[['nkill','nwound']].sum()
affects=affects.tail()
affects.reset_index(inplace=True)
affects
```

# **A bar plot on the number of people affected (killed and wonded) for the recent 5 years**

```python
affects.plot(x='iyear',y=['nkill','nwound'],kind='bar',figsize=(18,4));
plt.title('Affected no.of people')
plt.ylabel('No.of People');
```

# Inference: From the above bar plot we can see that the highest number of people wounded are in the year 2015 and the highest number of people who are killed are in the year 2014.

# **Count plot on the frequency of terrorism in each year**

```python
plt.figure(figsize = (15,7))
ax = sns.countplot(df['iyear'], palette= 'Blues');
plt.xticks(rotation = 90)
ax.set_xlabel('Year',fontsize=15)
ax.set_ylabel('Count', fontsize=15)
ax.set_title('Terrorism per Year',fontsize=15);
```

# **Inference:** From the above plot we can see that the maximum terrorism took place in the year 2014 and minimum terrorism took place in 1971
